```python
#importa as bibliotecas

import tensorflow as tf
import numpy as np
```

```python
#define o modelo

my_layer_1 = tf.keras.layers.Dense(units = 2, input_shape = [1])
my_layer_2 = tf.keras.layers.Dense(units = 1)
model = tf.keras.Sequential([my_layer_1, my_layer_2])
model.compile(optimizer = 'sgd', loss='mean_squared_error')
```

```python
#dados de entrada e saída

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

```python
#treino do modelo

model.fit(xs, ys, epochs = 500)
```

```python
#sumário do modelo

model.summary()
```

```python
#resultados

print(model.predict([10.0]))
print('\n')
print(my_layer_1.get_weights())
print('\n')
print(my_layer_2.get_weights())

w = model.get_weights()
b = model.get_weights()

print('w = ', w, 'b = ', b)
```

```python
# Exercicio de 3 camadas

# Importando as bibliotecas

import tensorflow as tf
import numpy as np

# Dados

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)

# Criação do modelo

my_layer_1 = tf.keras.layers.Dense (units = 2, input_shape = [1])
my_layer_2 = tf.keras.layers.Dense (units = 2, input_shape = [1])
my_layer_3 = tf.keras.layers.Dense (units = 1)
my_model = tf.keras.Sequential([my_layer_1, my_layer_2, my_layer_3])

# Compilação do modelo

my_model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

# Sumário do modelo

my_model.summary()

# Treinamento do modelo

my_model.fit(xs, ys, epochs = 500)

# Aplicação

print(my_layer_1.get_weights())
my_model.predict([10.0, 5.0, 5.5])
#my_model.save('my_model_trained.h5')
```
